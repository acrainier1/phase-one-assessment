```python
# answer problem 3 in this notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winemag10k.csv')
df.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
x = df.price
y = df.points

plt.scatter(x, y)
plt.title("Wines: Price vs. Points")
plt.xlabel("Price")
plt.ylabel("Points")
plt.savefig('wines_plot.png')
```
